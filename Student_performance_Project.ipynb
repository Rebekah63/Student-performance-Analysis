```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


#Load data set
grpdata = pd.read_csv("/content/drive/MyDrive/Students_Grading_Dataset(1).csv")


#Set index as Student_ID
grpdata.set_index('Student_ID', inplace=True)
grpdata.head()
```

```python
#Summary of the data
grpdata.info()
```

Missing values in columns: Attendance (%) ,Assignments_Avg, Parent_Education_Level  

```python
grpdata.describe()
```

```python
#check for duplicates
grpdata.duplicated().sum()
```

```python
#Check for missing values
grpdata.isna().sum()
```

```python
#Fill in missing values
grpdata['Parent_Education_Level'] = grpdata['Parent_Education_Level'].fillna('Other')
```

```python
grpdata['Assignments_Avg'].fillna(grpdata['Assignments_Avg'].median(), inplace=True)
grpdata['Attendance (%)'].fillna(grpdata['Attendance (%)'].median(), inplace=True)
```

```python
grpdata.isna().sum()
```

```python
#Replace CS with Computer Science
grpdata['Department'] = grpdata['Department'].replace('CS', 'Computer Science')
```

```python
grpdata.head()
```

No duplicates ,null values have been handled. Data Set is clean and ready for EDA

Data types : Categorical and numerical

```python
#numerical variables
num_cols = grpdata.select_dtypes('number').columns
```

EXPLORATORY DATA ANALYSIS

**UNIVARIATE ANALYSIS**

```python
#Visualize distribution Grade
plt.figure(figsize=(8, 6))
grpdata['Grade'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Blues'))
plt.title('Distribution of Grade')
plt.ylabel('')
plt.show()
```

Majority of the students(30%) achieve high scores (A)

**BIVARIATE ANALYSIS**

```python
#Parent_Education _Level by Grade

plt.figure(figsize=(10, 6))
sns.countplot(x='Parent_Education_Level', hue='Grade', data=grpdata, palette='Blues')
plt.title('Grouped Bar Chart: Parent_Education_Level vs Grade')
plt.show()
```

```python
#create a cross tab table
cross_tab = pd.crosstab(grpdata['Gender'], grpdata['Department'])
```

```python

# gender distribution across departments.
cross_tab.plot(kind='bar', stacked=True, colormap='Blues', figsize=(10,6))
plt.title('Gender distribution across departments')
plt.ylabel('Count')
plt.show()
```

**MULTIVARIATE ANALYSIS**

```python
#Visualize distribution of key numerical variables
plt.figure(figsize=(12, 6))
for i, num_cols in enumerate(['Study_Hours_per_Week','Attendance (%)','Midterm_Score', 'Final_Score', 'Total_Score']):
    plt.subplot(2, 3, i + 1)
    sns.histplot(grpdata[num_cols], kde=True)
plt.title('Distribution of key numerical variables across Grade')
plt.tight_layout()
plt.show()

```

```python
#correlation heatmap of numerical variables
plt.figure(figsize=(10, 8))
sns.heatmap(grpdata.select_dtypes('number').corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

Most correlations are close to 0, indicating very weak or no linear relationships among the numerical variables.


```python
plt.figure(figsize=(12, 6))

# Study Hours by Grade
plt.subplot(1, 2, 1)
sns.boxplot(x='Grade', y='Study_Hours_per_Week', data=grpdata, palette='Blues')
plt.title('Study Hours per Week by Grade')
plt.xlabel('Grade')
plt.ylabel('Study Hours per Week')

# Stress Level by Grade
plt.subplot(1, 2, 2)
sns.boxplot(x='Grade', y='Stress_Level (1-10)', data=grpdata, palette='viridis')
plt.title('Stress Level by Grade')
plt.xlabel('Grade')
plt.ylabel('Stress Level (1-10)')

plt.tight_layout()
plt.show()
```

**CHECK FOR OUTLIERS**

```python
# Set plot style
sns.set(style="whitegrid")

# List of continuous features
num_cols = [
    'Study_Hours_per_Week',
    'Attendance (%)',
    'Midterm_Score',
    'Final_Score',
    'Total_Score',
    'Assignments_Avg'
]

# Create subplots
plt.figure(figsize=(20, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i) # Adjusted subplot grid to fit 6 plots
    sns.boxplot(y=grpdata[col], data=grpdata, color='skyblue')
    plt.title(f'Box Plot of {col}')
    plt.xlabel('')
    plt.tight_layout()

plt.show()
```
